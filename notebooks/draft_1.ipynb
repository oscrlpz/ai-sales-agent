```python
import os

from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
DEEPSEEK_API_KEY = os.getenv("DEEPSEEK_API_KEY")
```

Mandar un mensaje a openai y recibir una respuesta

```python
AVAILABLE_MODELS = {
    "local-deepseek-r1:14b": {
        "model_name": "deepseek-r1:14b",
        "url": "http://localhost:11434/v1",
        "api_key": "not-needed",
    },
    "deepseek-v3": {
        "model_name": "deepseek-chat",
        "url": "https://api.deepseek.com",
        "api_key": DEEPSEEK_API_KEY,
    },
    "deepseek-r1": {
        "model_name": "deepseek-reasoner",
        "url": "https://api.deepseek.com",
        "api_key": DEEPSEEK_API_KEY,
    },
    "gpt-3.5-turbo": {
        "model_name": "gpt-3.5-turbo",
        "url": "https://api.openai.com/v1",
        "api_key": OPENAI_API_KEY,
    },
}
```

```python
model_name = "deepseek-v3"
model_info = AVAILABLE_MODELS[model_name]

opanai_client = OpenAI(
    api_key=model_info["api_key"], 
    base_url=model_info["url"],
)
model = model_info["model_name"]



chat_history = [
    {
        "role": "system",
        "content": "You are a pirate."
    },
    {
        "role": "user",
        "content": "Hello, who are you?"
    }
]
response = opanai_client.chat.completions.create(
    model=model,
    messages=chat_history,
)
response.choices[0].message.content
```

Luego sigue como mandar mensajes de whatsapp con twilio

```python
from twilio.rest import Client as TwilioClient

account_sid = os.getenv("TWILIO_ACCOUNT_SID")
auth_token = os.getenv("TWILIO_AUTH_TOKEN")
twilio_client = TwilioClient(account_sid, auth_token)

message = "hola"
message = twilio_client.messages.create(
    from_="whatsapp:+14155238886",
    to="whatsapp:+5215510535361",
    # to="whatsapp:+5215618581253",  # pep
    body=message,
    # content_sid="HXb5b62575e6e4ff6129ad7c8efe1f983e",
    # content_variables='{"1":"12/1","2":"3pm"}',
)

print(f"Message sent with SID: {message.sid}")
```

Y recuperar los mensajes de whatsapp con twilio

```python
messages = twilio_client.messages.list(
    to="whatsapp:+14155238886",
    # from_="whatsapp:+5215510535361",
)
for message in messages:
    print(f"Message SID: {message.sid}, Body: {message.body}, from: {message.from_}, to: {message.to}, date: {message.date_sent}")
```

```python
# list the name of all tha attributes of the message object
print("Attributes of the message object:")
for attr in dir(message):
    if not attr.startswith("__"):
        print(attr)
```

Ahor avoy a extraer la propuesta de valor de kavak

```python
import requests
from bs4 import BeautifulSoup
import tiktoken

url_kavak = "https://www.kavak.com/mx/blog/sedes-de-kavak-en-mexico"
```

```python
response = requests.get(url_kavak)
soup = BeautifulSoup(response.content, "html.parser")
# Find all the <h2> tags in the HTML
h2_tags = soup.find_all("h2")
# Print the text of each <h2> tag
plain_text = soup.get_text(separator="\n", strip=True)
print(plain_text)
```

```python
# save plaintext as a file
with open("/home/oscar/projects/ai-sales-agent/bot/config/kavak_info.txt", "w") as f:
    f.write(plain_text)
```

```python
def count_tokens(text):
    encoding = tiktoken.get_encoding("cl100k_base")  # Use for GPT-like models
    return len(encoding.encode(text))

count_tokens(plain_text)
```
